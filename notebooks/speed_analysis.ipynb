### Remove null, zero

```python
def remove_zero_columns(df):
    # Drop columns where all values are zero
    non_zero_columns = df.columns[(df != 0).any()]
    df_non_zero = df[non_zero_columns]
    return df_non_zero
```

```python
print('Before dropna df_stopped:',df_stopped.shape)
df_stopped = df_stopped.dropna()
print('After dropna df_stopped:',df_stopped.shape)
```

```python
print('Before dropna df_7hz:',df_7hz.shape)
df_7hz = df_7hz.dropna()
print('After dropna df_7hz:',df_7hz.shape)
```

```python
print('Before dropna df_10hz:',df_10hz.shape)
df_10hz = df_10hz.dropna()
print('After dropna df_10hz:',df_10hz.shape)
```

```python
print('Before dropna df_13hz:',df_13hz.shape)
df_13hz = df_13hz.dropna()
print('After dropna df_13hz:',df_13hz.shape)
```

```python
df_13hz.info()
```

### Add labels

```python
df_stopped['label'] = 0
df_stopped.head()
```

```python
df_7hz['label'] = 1
df_7hz.head()
```

```python
df_10hz['label'] = 2
df_10hz.head()
```

```python
df_13hz['label'] = 3
df_13hz.head()
```

### ETL

```python
df = pd.concat([df_stopped,df_7hz,df_10hz,df_13hz])
df
```

```python
DIM = 23
fontsize = 14

f = plt.figure(figsize=(DIM, DIM))
plt.matshow(df.corr(), fignum=f.number)
plt.xticks(range(df.select_dtypes(['number']).shape[1]), df.select_dtypes(['number']).columns, fontsize=fontsize, rotation=90)
plt.yticks(range(df.select_dtypes(['number']).shape[1]), df.select_dtypes(['number']).columns, fontsize=fontsize)
cb = plt.colorbar()
cb.ax.tick_params(labelsize=fontsize)
plt.title('Correlation Matrix', fontsize=16);

plt.savefig('corr.png')
```

```python
# shuffle data
df = df.sample(frac=1)
df.reset_index(drop=True, inplace=True)
df.head()
```

```python
y = df['label']
df.drop(columns=['label'], axis=1,inplace=True)
print(y.shape)
print(df.shape)
```

```python
df.head()
```

### Train model

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df, y, test_size=0.2, shuffle=True,random_state=42)
print('X_train',X_train.shape)
print('X_test',X_test.shape)
print('y_train',y_train.shape)
print('y_test',y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
```

```python
# Choose a Model
model = LogisticRegression()

# Train the Model
model.fit(X_train, y_train)
```

### Test model

```python
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)

print("Accuracy:", accuracy)
print("Classification Report:")
print(classification_report(y_test, y_pred))
```

### Export model

```python
from joblib import dump, load

dump(model, 'model.joblib') 
```

### Load and predict model

```python
saved_model = load('model.joblib') 
```

```python
# Predict input from dataframe
x = df.iloc[0]
print(saved_model.predict([x]), y[0])
```
